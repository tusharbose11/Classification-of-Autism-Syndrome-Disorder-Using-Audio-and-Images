# DATA AUGMENTATION

We will generate more data with the help of the library librosa. We will change the playback rate and vary the pitch to obtain new data as more data is required to train the CNN Model.

# Vary Time

```python
#time variation
rate = 0.81
path = "./updated/asd"
class_type = 'normal asd'.split()
while rate<=1.07:
    for filename in os.listdir(f'./updated/asd/'):
        if rate != 1.00:
            r1 = random.randint(0, 999999999)
            file_name = f'{filename}'
            y, sr = librosa.load('updated/asd/'+str(file_name), mono=True, duration=4)
            y_changed = librosa.effects.time_stretch(y, rate=rate)
            sf.write('updated/timevar/asd/'+str(r1)+str(file_name),y_changed, sr)
        else:
            pass
    rate=rate+0.01 
```

```python
path = "./updated"
file = open('finaldata1.csv', 'w', newline='')
with file:
    writer = csv.writer(file)
    writer.writerow(header)
class_type = ['asd']
#for asd
for filename in os.listdir(f'./updated/asd/'):
    file_name = f'{filename}'
    y, sr = librosa.load('updated/asd/'+str(file_name), mono=True, duration=4)
    chroma_stft = librosa.feature.chroma_stft(y=y, sr=sr)
    rms = librosa.feature.rms(y=y)
    spec_cent = librosa.feature.spectral_centroid(y=y, sr=sr)
    spec_bw = librosa.feature.spectral_bandwidth(y=y, sr=sr)
    rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr)
    zcr = librosa.feature.zero_crossing_rate(y)
    mfcc = librosa.feature.mfcc(y=y, sr=sr)
    to_append = f'{filename} {np.mean(chroma_stft)} {np.mean(rms)} {np.mean(spec_cent)} {np.mean(spec_bw)} {np.mean(rolloff)} {np.mean(zcr)}'    
    for e in mfcc:
        to_append += f' {np.mean(e)}'
    for g in class_type:
        to_append += f' {g}'
    file = open('finaldata1.csv', 'a', newline='')
    with file:
        writer = csv.writer(file)
        writer.writerow(to_append.split())
```

```python
data = pd.read_csv('finaldata.csv')
data.shape
```

```python
# Dropping unneccesary columns
data = data.drop(['filename'],axis=1)
data.head()
```

```python
data = data.sample(frac=1).reset_index(drop=True)
data
```

```python
class_types = data.iloc[:, -1]
encoder = LabelEncoder()
y = encoder.fit_transform(class_types)
print(y)
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(np.array(data.iloc[:, :-1], dtype = float))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)
```

```python
model = models.Sequential()
model.add(layers.Dense(256, activation='relu', input_shape=(X_train.shape[1],)))
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(2, activation='softmax'))
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
history = model.fit(X_train,
                    y_train,
                    epochs=20,
                    batch_size=128)
```

```python
test_loss, test_acc = model.evaluate(X_test,y_test)
print('test_acc: ',test_acc)
```
